```python
import os
```

```python
gsutil_cmd = "gsutil ls gs://cs327e-open-access-2/school-enrollments/*"
file_listings = os.popen(gsutil_cmd)
dataset="school_enrollments"

for file in file_listings:

    start_index = file.rindex("_") + 1
    end_index = file.rindex(".")
    table = file[start_index:end_index]
    
    if table in ("co", "district", "fl", "ga", "id", "ma", "mo", "ne", "ny", "or", "school", "sd", "wi"):
        print("skipping " + table)
        continue

    bq_cmd = "bq --location=US load --autodetect --skip_leading_rows=1 "\
             "--source_format=CSV " + dataset + "." + table + ' ' + file
    print(bq_cmd)
    
    os.system(bq_cmd)
```

Query #1 This query searches from the schools in districts that have'DALLAS' anywhere in the district name where the total value has an entry and the school is named. The query is grouped by school, grade, year. The returned attributes is school, grade, year and the sum of the total for the school during the specific year and for each grade.

```python
%%bigquery
select school, grade, year, sum(cast(replace(replace(total, ",", ""), "<", "") as int64)) as total
from school_enrollments.tx
where total != "" and school != "" and district like '%DALLAS%'
group by school, grade, year
order by total desc
```

Query #2 This query searches from the schools in districts that have'HOUSTON' anywhere in the district name where the total value has an entry and the school is named. The having clause requires that the total is greater than or equal to 0. The query is grouped by school, grade, year. The returned attributes is school, grade, year and the sum of the total for the school during the specific year and for each grade.

```python
%%bigquery
select school, grade, year, sum(cast(replace(replace(total, ",", ""), "<", "") as int64)) as total
from school_enrollments.tx
where total != "" and school != "" and district like '%HOUSTON%'
group by school, grade, year
having total >= 0
order by total desc
```

Query #3 This query merges two queries into one table. Each query returns the distract, grade, and count of how many schools in that district have these grade levels. The queries are then grouped by the district name and grade levels. The return table is ordered by grade in ascending alphabetical order.

```python
%%bigquery
select district, grade, count(school) as Schools_with_grade
from school_enrollments.ca
where district like "Chico Unified"
group by district, grade
union distinct
select district, grade, count(school) as School_with_grade
from school_enrollments.ca
where district like "Paradise Unified"
group by district, grade
order by grade;
```

Query #4 This query searches all of the schools in California for school enrollment in 2020 for the total number of students enrolled in all the grades combined. The results are ordered by the total number of students in descending order.

```python
%%bigquery
select school, year, sum(total) as Total_students
from school_enrollments.ca
where school != "" and total != 0
group by school, year, total
having year = 2020
order by total desc
```

Query #5 This query selects the district, year, total, white, black, hispanic, native american alaskan native, asian, native hawaiian pacific islander, pacific islander, multiracial, unknown race, and Total Students from the entire school enrollment in california.  The query limits the selection to those with total students greater than 0 and orders by district.

```python
%%bigquery
select district, year, total, white,black,hispanic,native_american_alaskan_native,asian,native_hawaiian_pacific_islander,pacific_islander,multiracial, unknown_race, sum(total) as Total_students
from school_enrollments.ca
where school != ""
group by district, year, total, white,black,hispanic,native_american,native_american_alaskan_native,asian,native_hawaiian,asian_pacific_islander,native_hawaiian_pacific_islander,pacific_islander,multiracial, unknown_race
having total > 0
order by district
```

Query #6 This query picks out the school, number of homeless, low income students as well as the total students at the school for schools in Washington state.  The results are limited to those schools with less than or equal to 50 low income students and ordered by the name of the school.

```python
%%bigquery
select school, homeless, low_income, sum(total) as Total_students
from school_enrollments.wa
where school != ""
group by school,low_income, homeless
having count(low_income) <= 50
order by school
```

Query #7 This query uses the schools as well as the grade levels and total female and male student population of schools in california.  The results are limited in schools where the number of females is less than the number of males and is ordered by school name ascending.

```python
%%bigquery
select school, grade, sum(female) as females, sum(male) as males
from school_enrollments.ca
where school != ""
group by school,grade, female, male
having female < male
order by school
```

Query #8 This query joins schools from the Fremont Unified and Dublin Unified districts by the grade level and sees if there are equal amounts of schools with those grade levels listed.

```python
%%bigquery
select grade, count(school) as Schools_with_grade
from school_enrollments.ca
where district like "Fremont Unified"
group by grade
union all
select grade, count(school) as Schools_with_grade
from school_enrollments.ca
where district like "Dublin Unified"
group by grade
order by grade;
```
