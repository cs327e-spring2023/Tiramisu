Inner join rental and customer tables by customer_id, filtering rows where the first name begins with A and orders by the customer_id.

```python
!psql -c "SELECT a.rental_id, a.customer_ID, b.first_name, b.last_name FROM pagila.rental a INNER JOIN pagila.customer b ON a.customer_ID = b.customer_ID WHERE b.first_name LIKE 'A%' ORDER BY a.customer_ID ASC LIMIT 10"
```

Inner join staff and store tables by store_id, filtering rows where the active status is 't' and orders by the store_id.

```python
!psql -c "SELECT a.active, a.staff_id, a.first_name, a.last_name, b.store_id, b.manager_staff_id FROM pagila.staff a INNER JOIN pagila.store b ON a.store_id = b.store_id WHERE a.active ='t' ORDER BY a.store_id LIMIT 10"
```

Inner join film_actor and actor tables by actor_id, filtering rows where the actor_id is greater than 10 and orders by the actor_id.

```python
!psql -c "SELECT a.film_id, a.actor_id, b.first_name, b.last_name FROM pagila.film_actor a INNER JOIN pagila.actor b ON a.actor_id = b.actor_id WHERE a.actor_id > 10 ORDER BY a.actor_id LIMIT 10"
```

Right outer join customer and payment tables by the customer_id, filtering rows where the amount is greater than 5 ordered by customer_id in alphabetical order.

```python
!psql -c "SELECT a.customer_id, a.first_name, a.last_name, b.payment_id, b.rental_id, b.amount FROM pagila.customer a RIGHT OUTER JOIN pagila.payment b ON a.customer_id = b.customer_id WHERE b.amount > 5 ORDER BY a.customer_id LIMIT 10"
```

Left outer join film and language tables by the language_id, filtering rows where the length is less than 5 ordered by title in alphabetical order.

```python
!psql -c "SELECT a.film_id, a.title, a.length, b.language_id, b.name FROM pagila.film a LEFT OUTER JOIN pagila.language b ON a.language_id = b.language_id WHERE a.length < 50  ORDER BY a.title LIMIT 10"
```

Right outer join address and city tables by the customer_id, filtering rows where the postal code begins with 7 ordered by the city_id in alphabetical order.

```python
!psql -c "SELECT a.postal_code, a.city_id, b.city FROM pagila.address a RIGHT OUTER JOIN pagila.city b ON a.city_id = b.city_id WHERE a.postal_code LIKE '7%' ORDER BY a.city_id LIMIT 10"
```

Self join film_actor table to itself, filtering rows where the actor_id doesn't equal to itself but the film_id is equal and orders by the actor_id.

```python
!psql -c "SELECT A.actor_id, B.actor_id, B.film_id FROM pagila.film_actor A, pagila.film_actor B WHERE A.actor_id <> B.actor_id AND A.film_id = B.film_id ORDER BY A.actor_id LIMIT 10"
```
